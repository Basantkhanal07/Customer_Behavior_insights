```python
# 02_exploratory_data_analysis.ipynb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned data
df = pd.read_csv('../data/customer_behavior_clean.csv')
```

```python
# Descriptive statistics
print(df[['age','annual_income','spending_score']].describe())
```

```python
# --- Histograms ---
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Age Distribution")

plt.subplot(1,2,2)
sns.histplot(df['annual_income'], bins=30, kde=True)
plt.title("Annual Income Distribution")
plt.show()
```

```python
# --- Scatterplot ---
plt.figure(figsize=(6,5))
sns.scatterplot(x='annual_income', y='spending_score', data=df)
plt.title("Annual Income vs Spending Score")
plt.show()
```

```python
# --- Bar plot: PurchaseHistory ---
plt.figure(figsize=(6,4))
sns.countplot(y='purchase_history', data=df,
              order=df['purchase_history'].value_counts().index)
plt.title("Purchase History by Category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```
